```python
def find_promoted_employees(daishin_executive, daishin_Annual_Financial_Info):
    increased_years = []  # 매출이 증가한 연도를 담을 리스트
    promoted_employees = []  # 승진 또는 입사한 인물을 담을 리스트

    # 매출액이 증가한 연도 찾기
    for index, row in daishin_Annual_Financial_Info.iterrows():
        if row['이전 연도 대비 증가량'] > 0:
            increased_years.append(row['연도'])

    for year in increased_years:
        for index, row in daishin_executive.iterrows():
            if row['보고서연도'] == year:
                previous_year = year - 1

                # 직위 변화 확인
                previous_row = daishin_executive.loc[daishin_executive['보고서연도'] == previous_year]
                if len(previous_row) > 0 and row['직위'] != previous_row['직위'].values[0]:
                    promoted_employees.append(row['이름'])

                # 입사 여부 확인
                start_year = int(row['재직기간'].split('.')[0])
                if start_year == year:
                    promoted_employees.append(row['이름'])

    return promoted_employees
```

```python
promoted_employees = find_promoted_employees(daishin_executive, daishin_Annual_Financial_Info)

print('매출액이 증가한 연도에 승진이나 입사한 인물:')
for employee in promoted_employees:
    print(employee)
```

```python
def find_increased_stockholders(daishin_shareholder, daishin_Annual_Financial_Info):
    increased_years = []  # 매출이 증가한 연도를 담을 리스트
    increased_stockholders = []  # 주식보유량이 증가한 인물을 담을 리스트

    # 매출액이 증가한 연도 찾기
    for index, row in daishin_Annual_Financial_Info.iterrows():
        if int(row['이전 연도 대비 증가량']) > 0:
            increased_years.append(row['연도'])

    for year in increased_years:
        for index, row in daishin_shareholder.iterrows():
            if row['보고서연도'] == year:
                # 주식보유량 변화 확인
                initial_stock = row['기초주식수']  # 매출이 증가한 연도의 기초주식수
                final_stock = row['기말주식수']  # 매출이 증가한 연도의 기말주식수
                stock_change = final_stock - initial_stock  # 주식보유량 변화

                if stock_change > 0:
                    increased_stockholders.append(row['이름'])

    return increased_stockholders
```

```python
increased_stockholders = find_increased_stockholders(daishin_shareholder, daishin_Annual_Financial_Info)

print('주식보유량이 증가한 인물:')
for employee in increased_stockholders:
    print(employee)
```

```python
combined_list_temp = promoted_employees + increased_stockholders
```

```python
counter = Counter(combined_list_temp)

# 등장 횟수가 많은 순서대로 정렬된 튜플 리스트 생성
sorted_elements = sorted(counter.items(), key=lambda x: x[1], reverse=True)

# 정렬된 리스트에서 요소와 등장 횟수 출력
for element, count in sorted_elements:
    print(f'{element}: {count}')
```
